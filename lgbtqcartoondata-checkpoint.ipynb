```python
import pandas

pandas.options.display.max_columns = None
pandas.options.display.max_rows = None

file_data = pandas.read_csv("lgbtq.csv")
file_data
```

```python
"""What audience is most likely to see LGBTQ+ representation?
- Find the unique age groups
- Count number of shows in each age group
- Compare number of shows to see which one is highest
"""
# find all age ratings in data base
age_ratings = file_data['TV_rating'].dropna().unique()
age_ratings

TV_Y7 = 0
TV_PG = 0
TV_14 = 0
TV_MA = 0
TV_G = 0
TV_Y = 0

for index, row in file_data.iterrows():
    if row['TV_rating'] == "TV-Y7":
        TV_Y7 += 1
    if row['TV_rating'] == "TV-PG":
        TV_PG += 1
    if row['TV_rating'] == "TV-14":
        TV_14 += 1
    if row['TV_rating'] == "TV-MA":
        TV_MA += 1
    if row['TV_rating'] == "TV-G":
        TV_G += 1
    if row['TV_rating'] == "TV-Y":
        TV_Y += 1     
print("TV-Y7 =", TV_Y7)
print("TV-Y =", TV_Y)
print("TV-G =", TV_G)
print("TV-PG =", TV_PG)
print("TV-14 =", TV_14)
print("TV-MA =", TV_MA)
```

```python
"""
Do prominent networks represent LGBTQ+ characters?
- Find all the unique networks
- Take the ones that most people watch (based on preliminary scroll through data)
- Compare numbers to find which one is the highest
"""
# find all networks in data base
network = file_data['network'].dropna().unique()
network

netflix = 0
hulu = 0
cartoon_network = 0
amazon_prime = 0
hbo_max = 0
adult_swim = 0
youtube = 0

for index, row in file_data.iterrows():
    if row['network'] == "Netflix":
        netflix += 1
    if row['network'] == "Hulu":
        hulu += 1
    if row['network'] == "Cartoon Network":
        cartoon_network += 1
    if row['network'] == "Amazon Prime":
        amazon_prime += 1
    if row['network'] == "HBO Max":
        hbo_max += 1
    if row['network'] == "Adult Swim":
        adult_swim += 1
    if row['network'] == "YouTube":
        youtube += 1
print("Netflix =", netflix)
print("Hulu =", hulu)
print("Cartoon Network =", cartoon_network)
print("Amazon Prime =", amazon_prime)
print("HBO Max =", hbo_max)
print("Adult Swim =", adult_swim)
print("YouTube =", youtube)
```
